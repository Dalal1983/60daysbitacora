# Carga de las librerías

```python
from sklearn.datasets import load_iris
from sklearn import tree
from matplotlib import pyplot as plt
import pandas as pd
```

# Dataset

X: Corresponde a los atributos que se consideran son los que puden revelar más precisamente la clasificación de las flores.
* X[0] : largo de sépalo
* X[1] : ancho de sépalo
* X[2] : largo de pétalo
* X[3] : ancho de pétalo

y: corresponde a las distintas clasificaciones que puede tener la flor según la muestra
* 0 = Setosa
* 1 = Versicolor
* 2 = Virginica

```python
iris = load_iris()
```

# Asignación del modelo

```python
clf = tree.DecisionTreeClassifier(criterion='gini')
clf = clf.fit(iris.data, iris.target)
```

# Predicción de nuevos valores

```python
plt.figure(figsize=(20,10))

tree.plot_tree(clf)
plt.show()
```

# Predicción

```python
clf.predict([[5., 2., 4., 3.]])
```
